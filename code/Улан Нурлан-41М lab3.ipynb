```python
# @title 2. Проверьте набор данных
import pandas as pd

# Правильная ссылка на оригинальный файл в репозитории GitHub
file_url = "https://raw.githubusercontent.com/AI-is-out-there/data2lab/main/%D0%BC%D0%BE%D0%B4%D1%83%D0%BB%D1%8C%203%20-%20%D0%B4%D0%B0%D1%82%D0%B0%D1%81%D0%B5%D1%82%20-%20%D0%BF%D1%80%D0%B0%D0%BA%D1%82%D0%B8%D0%BA%D0%B0.csv"

# 2.Список столбцов для выбора
selected_columns = ['Count_subj', 'rr_interval', 'p_end', 'qrs_onset', 'qrs_end', 'p_axis', 'qrs_axis', 't_axis', 'Healthy_Status']

# 1.Загрузить первые 5000 строк

df = pd.read_csv(file_url, nrows=5000)

# Выберить необходимые столбцы
df = df[selected_columns]

print("Набор данных был успешно загружен и подготовлен.")
print(f"Количество загруженных строк: {len(df)}")
print("Первые 5 строк набора данных:")
print(df.head())
print("\nИнформация о наборе данных:")
df.info()
```

```python
# @title 3. Запустить модель
from pycaret.classification import *

# Инициализируйте среду PyCaret
clf1 = setup(data=df,
             target='Healthy_Status',
             session_id=123,
             normalize=True,   # Нормализовать характеристики
             fix_imbalance=True # Попробуйте исправить дисбаланс классов.
            )

# Сравнение моделей (отсортировано по показателю F1)
print("\nСравнение моделей с использованием PyCaret (сортировка по F1)...")
best_model = compare_models(sort='F1')

print(f"\nЛучшая модель, выбранная по метрике F1: {type(best_model).__name__}")

# Оцените лучшую модель на отложенном тестовом наборе
print("\nОцените лучшую модель на отложенном тестовом наборе")
evaluation_results = predict_model(best_model)

print("\nМиссия выполнена.")
```
